```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Rosenbrock_tau.csv')
fig, ax = plt.subplots()
fig.set_size_inches(10, 4)
sns.boxplot(x='tau', y='epoch_stop', data=df, 
            palette=sns.color_palette("blend:#7AB,#EDA", 10), width=0.4, ax=ax)
plt.xlabel('Tau', size = 14, **{'fontname':'Times New Roman'})
plt.ylim(0, 500)
plt.ylabel('Epoch Stop', size = 14, **{'fontname':'Times New Roman'})
plt.show()
```

```python
df = pd.read_csv('500/R_tau_lower_phi_1.csv')
fig, ax = plt.subplots()
fig.set_size_inches(10, 4)
sns.boxplot(x='tau', y='fitness', data=df, 
            palette=sns.color_palette("blend:#7AB,#EDA", 10), width=0.4, ax=ax)
plt.xlabel('Tau', size = 14, **{'fontname':'Times New Roman'})
plt.ylabel('Fitness', size = 14, **{'fontname':'Times New Roman'})
#plt.ylim(0, 0.0000000000000000000000000001)
# make the y axis log scale
ax.set_yscale('log')
plt.show()
```
