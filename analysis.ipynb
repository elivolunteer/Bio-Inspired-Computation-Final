```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('500/R_tau_lower_phi_1.csv')
fig, ax = plt.subplots()
fig.set_size_inches(10, 4)
sns.boxplot(x='tau', y='fitness', data=df, 
            palette=sns.color_palette("blend:#7AB,#EDA", 10), width=0.4, ax=ax)
plt.xlabel('Tau', size = 14, **{'fontname':'Times New Roman'})
plt.ylabel('Fitness', size = 14, **{'fontname':'Times New Roman'})
# make the y axis log scale
ax.set_yscale('log')
#plt.title('Rosenbrock Function Varrying Tau 50 Runs Each, Every Tau Decreases Cog by 0.1', size = 14, **{'fontname':'Times New Roman'})
# Ignore the outliers
plt.show()
```
