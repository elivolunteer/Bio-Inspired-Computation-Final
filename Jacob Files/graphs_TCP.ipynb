```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 15}

plt.rc('font', **font)

# Modifying Particles

#for each particle value
#determine the range of epoch_stop values that did converge

#read rosenbrock file to df
rosenbrock_df = pd.read_csv("TCP__Rosenbrock.csv")

#separate into 4 df based on parameters
#10 * 20 tests = 200 tests

#particles_df = rosenbrock_df[0:200]
#converged = particles_df[particles_df['fitness'] <= 1e-10]
```

- Display Particle Boxplot

```python
particle_df = rosenbrock_df[0:75].reset_index()
converged_particle = []
x_ticks = []
epochs = []

for i in range(0,75):
    if particle_df.loc[i,'fitness'] <= 1e-10:
        epochs.append(particle_df.loc[i,'epoch_stop'])

print(f"epoch length: {len(epochs)}")

if len(epochs) > 0:
        print("hi")
        x_ticks.append(particle_df.loc[i,'particles'])
        converged_particle.append(epochs)

particle_means = []
for j in converged_particle:
    if len(j) != 0:
        particle_means.append(sum(j)/len(j))

print(f"There are {len(epochs)} converged particles.")
print(f"The mean convergence time is {particle_means}")

plt.figure(figsize=(15,10))
plt.boxplot(converged_particle,positions=range(0,len(particle_means)),showmeans=True)
plt.plot(particle_means)
plt.gca().axes.xaxis.set_ticklabels(x_ticks)
plt.xlabel("With Dynamic Phi_2")
plt.tick_params(axis='x',which='both',bottom=False,top=False,labelbottom=False)
plt.ylabel("Epoch Stop")
plt.title("Epoch Stop (of converging tests) vs. No. of Particles (Rosenbrock_TCP)")
plt.show()
```
